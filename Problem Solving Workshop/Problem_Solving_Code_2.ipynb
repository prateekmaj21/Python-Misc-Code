```python
import matplotlib.pyplot as plt
import numpy as np

def calculate_cagr(initial_value, final_value, num_periods):
    """Calculates the Compound Annual Growth Rate (CAGR)."""
    return (final_value / initial_value)**(1 / num_periods) - 1

# Sample data (replace with your actual data)
# Assuming sales are for the first 5 months and we want to predict June.
sales = [50000, 52000, 48000, 55000, 60000]
months = np.arange(1, len(sales) + 1)

# Calculate CAGR for the first 5 months
cagr = calculate_cagr(sales[0], sales[-1], len(sales)-1)


# Project sales for June (month 6)
projected_sales_june = sales[-1] * (1 + cagr)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(months, sales, marker='o', linestyle='-', label='sales (First 5 Months)')
plt.plot(6, projected_sales_june, marker='x', markersize=10, color='green', label='Projected Sales for June')


plt.xlabel("Month")
plt.ylabel("SALES")
plt.title("SALES Projection for June using CAGR")
plt.xticks(np.append(months,6))
plt.legend()
plt.grid(True)
plt.show()
```
